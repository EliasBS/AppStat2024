# Central Limit Theorem (CLT)

The aim of this notebook is to illustrate the Central Limit Theorem (CLT) through concrete examples.

When you add random numbers from different distributions together and plot the distribution of these sums, you end up with a Gaussian distribution, as dictated by the CLT.
The example also illustrates how widths (and therefore uncertainties) are added in quadrature, as one has to divide the sum by the square root of the number of random numbers that went into the sum in order to get a Gaussian of unit width (when using random numbers of unit width, i.e. RMSE $= \sigma = 1$).

### Distributions to sample from:

Here we set the parameters for the experiement. We are going to play around with these more.
For now we choose to take __10 uniform numbers__ and consider their sum __1000 times__. 

```python
N_experiments = 1000                  # Number of sums produced
N_uniform     = 10                    # Number of uniform numbers used in sum
N_exponential = 10                     # Number of exponential numbers used in sum
N_cauchy      = 0                     # Number of cauchy numbers used in sum
```

## Counters to put simulated values into:

We initializing some zero-filled numpy arrays and also a counter to count how many of the produced sums that fall outside some range ($\pm 3\sigma$).

```python
x_uniform = np.zeros((N_uniform, N_experiments))
x_exponential = np.zeros((N_exponential, N_experiments))
x_cauchy = np.zeros((N_cauchy, N_experiments))
x_sum = np.zeros((N_experiments))

N3_sigma = 0        # Counter for the number of produced sums, that fall outside +-3 sigma
```

Next, we loop over "experiments" to make the code more explicit. This could in principle be avoided (which typically also makes the code faster), but for now we keep it, also to be able to comment on what goes on.

```python
for iexp in range(N_experiments): 

    if ((iexp+1) % 500 == 0): 
        print("At iexp : ", iexp+1)         # Show progress!
    sum_value = 0.0                         # sum_value is the number we are going to add random numbers to!
                                            # According to the CLT, it should be Gaussianly distributed.
 
    # Generating uniform numbers (with mean 0, and RMS of 1):
    x_uniform[:, iexp] = (np.random.uniform(size=N_uniform) - 0.5) * np.sqrt(12)   # ...sqrt(12). Why? Possibly see references above!
# ...sqrt(12). Why? Possibly see references above!
#Why is it square root of 12? Because the variance of a uniform distribution is 1/12*(b-a)^2, 
# where a and b are the limits of the distribution. In this case, a=-0.5 and b=0.5, so the variance is 
# 1/12*(0.5-(-0.5))^2 = 1/12*1^2 = 1/12. The standard deviation is the square root of the variance, 
# so the standard deviation is sqrt(1/12) = 1/sqrt(12). 
# The factor of 1/sqrt(12) is then multiplied by the random numbers to get the standard deviation of 1.

    sum_value += x_uniform[:, iexp].sum()               # We add the numbers in "sum_value"

    # Generating exponential numbers (with mean 0, and RMS of 1):
    x_exponential[:, iexp] = np.random.exponential(size=N_exponential) - 1   
    # Exponential starting at -1. Why? 
    # The exponential distribution is defined for x>0, so we need to shift it to the left by 1 to get it to start at -1.

    sum_value += x_exponential[:, iexp].sum()

    # Generating numbers according to a Cauchy distribution (1 / (1 + x^2)):
    x_cauchy[:, iexp] = np.random.standard_cauchy(size=N_cauchy)
    x_cauchy[:, iexp] = x_cauchy[:, iexp] - x_cauchy[:, iexp].mean()
    
    #Why is the normalized? 

    #The Cauchy distribution is a probability distribution with infinite variance, so it doesn't have a well-defined mean or standard deviation.


    sum_value += x_cauchy[:, iexp].sum()
  


    N_total = N_uniform + N_exponential + N_cauchy
    sum_value = sum_value / np.sqrt(N_total)              
    # Ask yourself, why I divide by sqrt(N)?
    # The sum of the random numbers is a sum of N random variables.
    # According to the Central Limit Theorem, the sum of N random variables will be Gaussian distributed with mean N*mean and variance N*variance.
    x_sum[iexp] = sum_value

    # Count the number of times, that the sum is not within 3 sigma of a unit Gaussian:
    if not (-3.0 < sum_value < 3.0):
        N3_sigma += 1
```

Now we have created three 2-dimensional arrays of shape (N_pdf, N_experiments), e.g. (10, 1000). We now flatten the arrays to get 1D arrays in order to see their distribution:

```python
x_uniform = x_uniform.flatten()
x_exponential = x_exponential.flatten()
x_cauchy = x_cauchy.flatten()



empty = np.zeros((5,5))
empty[:, 1] = (np.random.uniform(size=5))
print(empty)
```

## Draw output plots with unit Gaussian for reference:

Having visualized the individual distributions above, we now visualize the sum of the different distributions, `x_sum`. We first plot a histogram of it in the range from `xmin` to `xmax`, and overlay it with a unit Gaussian (scaled to match histogram normalisation).

```python
# Define the Gaussian:
def gauss_pdf(x, N, mu, sigma):
    return N / np.sqrt(2 * np.pi) / sigma * np.exp(-0.5 * (x - mu) ** 2 / sigma ** 2)
```

```python
xmin, xmax = -5, 5

fig2, ax2 = plt.subplots(figsize=(16, 7)) 
hist2 = ax2.hist(x_sum, bins=N_bins, range=(xmin, xmax), histtype='step', color="black", linewidth=2, label="Data")
ax2.set(xlabel='Sum', ylabel='Frequency / 0.1', title='Histogram of x_sum (sum of input values)')

binwidth = (xmax-xmin) / N_bins             # The scale factor between histogram and the fit is the bin width
x_gauss = np.linspace(xmin, xmax, 1000)     # Create the x-axis for the plot of the fitted function
y_gauss_unit = binwidth * gauss_pdf(x_gauss, len(x_sum), 0, 1) # Unit Gaussian
y_gauss_dist = binwidth * gauss_pdf(x_gauss, len(x_sum), np.mean(x_sum), np.std(x_sum, ddof = 1)) 
ax2.plot(x_gauss, y_gauss_unit, '-', color='blue', label='Unit Gauss (no fit)', lw = 4)
ax2.plot(x_gauss, y_gauss_dist, '-', color='orange', label='Distribution Gauss (mean and std)', lw = 4)
ax2.legend(loc='upper right', fontsize=20)
fig2.tight_layout()
if save_plots :
    fig2.savefig('Histogram.pdf', dpi=600)
```

Do you understand why the uniform distribution needs to go from $\pm \sqrt 3$ in order to give a distribution with a width of $1$ (i.e. unit) and why you subtract $1$ from the exponential distribution (and how this works at all)?

The uniform distribution needs to go from (-\sqrt{3}) to (\sqrt{3}) in order to give a distribution with a width (standard deviation) of 1 because of the properties of the uniform distribution.

For a uniform distribution (U(a, b)), the standard deviation (\sigma) is given by:

$[ \sigma = \frac{b - a}{\sqrt{12}} ]$

In this case, we want the standard deviation to be 1. Let's set $(a = -\sqrt{3})$ and $(b = \sqrt{3})$. Plugging these values into the formula, we get:

$[ \sigma = \frac{\sqrt{3} - (-\sqrt{3})}{\sqrt{12}} = \frac{2\sqrt{3}}{\sqrt{12}} = \frac{2\sqrt{3}}{2\sqrt{3}} = 1 ]$

Thus, the uniform distribution $(U(-\sqrt{3}, \sqrt{3}))$ has a standard deviation of 1. This is why the uniform distribution needs to go from $(-\sqrt{3})$ to $(\sqrt{3})$ to achieve the desired width.

1. What is the mean and RMSE (i.e. Standard Deviation) of the input distributions?

The mean of an input distribution is the average value of all the data points in that distribution. It is calculated by summing all the values and then dividing by the number of values.

The RMSE (Root Mean Square Error), which is equivalent to the standard deviation in this context, measures the amount of variation or dispersion of the data points from the mean. It is calculated by taking the square root of the average of the squared differences between each data point and the mean.

In summary:

Mean: The average value of the data points.
RMSE (Standard Deviation): A measure of the spread of the data points around the mean.


2. Why is there a $\frac{1}{\sqrt N}$ at the bottom of cell [7] in the line `sum_value = sum_value / np.sqrt(Ntotal)` (when summing up the various contributions to sum)?
    Hint: Assume that I always wanted to compare the distribution of sums with a UNIT Gaussian.
The term $\frac{1}{\sqrt{N}}$ in the line sum_value = sum_value / np.sqrt(Ntotal) is related to the Central Limit Theorem (CLT) and the normalization of the sum of random variables.

Central Limit Theorem (CLT)
The CLT states that the sum (or average) of a large number of independent and identically distributed (i.i.d.) random variables, each with finite mean and variance, will be approximately normally distributed. This normal distribution has a mean equal to the sum of the means of the individual variables and a variance equal to the sum of their variances.

Normalization
When summing up random variables, the variance of the sum increases with the number of variables. Specifically, if you have (N) i.i.d. random variables each with variance (\sigma^2), the variance of their sum is (N\sigma^2). To normalize this sum so that it has a constant variance (typically 1), you divide by (\sqrt{N}).

In Your Code
In the line sum_value = sum_value / np.sqrt(Ntotal), Ntotal likely represents the total number of random variables being summed. By dividing sum_value by (\sqrt{Ntotal}), you are normalizing the sum so that its variance remains constant, which is a common step when applying the CLT to ensure that the resulting distribution is standard normal (mean 0, variance 1).

This normalization is crucial for comparing sums of different numbers of random variables on the same scale.

Example
If you sum (N) random variables (X_i) with mean (\mu) and variance (\sigma^2), the sum (S_N = \sum_{i=1}^N X_i) has mean (N\mu) and variance (N\sigma^2). To standardize this sum, you subtract the mean and divide by the standard deviation:

$Z = \frac{S_N - N\mu}{\sqrt{N\sigma^2}} $

In your code, the division by $(\sqrt{Ntotal})$ is part of this standardization process
